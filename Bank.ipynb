## Marketing Introduction:
Marketing is the process by which companies/businesses create value for customers and build strong customer relationships in order to capture value from customers in return.

**Marketing campaigns** are characterized by focusing on customer needs and their overall satisfaction. There are variables that need to be taken into consideration when making a marketing campaign

1. Segment of the Population: To which segment of the population is the marketing campaign going to and why:
2. Distribution channel to reach the customer's place: Which instrument should be used to get the message out e.g. Social Media, billboards, TV etc.
3. Price: What's the best price to offer to potential clients
4. Promotional Strategy: This is the way the strategy is going to be implemented.




The **end goal** of using this dataset is generally to **predict customer behavior** and **optimize marketing strategies for a company**.

# Preprocessing

**Objective**
1. Identifying Unique values in Categorical columns
2. Encoding Categorical values into numerical values
3. Analyse the data for Linearity and Outliers


```python
dataframe['y']=(dataframe['y']=='yes').astype(int)
# Encoding the target values, 'yes' = 1 and 'no' = 0
```

```python
obj_col = dataframe.select_dtypes(include='object').columns

print(obj_col)
```

```python
for col in obj_col:
    print(f"column: {col}")
    print(dataframe[col].unique())
```

```python
dataframe['default'] = (dataframe['default']=='yes').astype(int)
dataframe['housing'] = (dataframe['housing']=='yes').astype(int)
dataframe['loan'] = (dataframe['loan']=='yes').astype(int)

obj_col.drop(['default', 'loan', 'housing'])
```

```python
for col in obj_col:
    print(f'-----{col}-------')
    print(dataframe[col].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

Le = LabelEncoder()

dataframe_2 = dataframe.copy()

for col in obj_col:
    dataframe_2[col] = Le.fit_transform(dataframe[col])
```

```python
dataframe_2.head()
```

We have now encoded our categorical data into numerical values. This is important because machine learning algorithms are mathematical procedures that rely solely on numerical data, they can't process raw text.

```python
for col in obj_col:
    print(f"column: {col}")
    print(dataframe_2[col].unique())
```

```python
print(dataframe["y"])
```

```python
import matplotlib.pyplot as plt

for label in dataframe[:-1]:
    plt.hist(dataframe[dataframe["y"] == 1][label], color='blue', label='yes', alpha=0.7, density=True)
    plt.hist(dataframe[dataframe["y"] == 0][label], color='red' , label='no', alpha=0.7, density=True)
    plt.title(label)
    plt.ylabel("Probability")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```

Majority of 'default' values are zero, so we can delete it

```python
dataframe_2.drop(['default'], axis=1, inplace=True)
dataframe_2.head()
```

```python
num_col = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
```

```python
import seaborn as sns

plt.figure(figsize=(20, 60), facecolor='white')
plotnumber = 1

for col in num_col:
    ax = plt.subplot(12, 3, plotnumber)
    sns.boxplot(dataframe_2[col])
    plt.xlabel(col)
    plotnumber += 1
plt.show()
```
